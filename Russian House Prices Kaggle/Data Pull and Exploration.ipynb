### Pick only certain columns for now

```python
train_data_select = train_data[['price_doc','id','timestamp','full_sq','floor','max_floor','material','build_year','num_room','kitch_sq','state','product_type','sub_area']]
```

```python
print 'Unique values for max_floor: \n', train_data_select['max_floor'].value_counts(dropna = False)
```

```python
#Automating the above code for all columns using lists
columns = list(train_data_select.columns)

for i in columns:
    print 'Unique values for', i ,': \n', train_data_select[i].value_counts(dropna = False)
```

```python
train_data_select.dtypes
```

```python
#convert Date column to Date from object dtype
train_data_select =  train_data_select.assign(Date =  pd.to_datetime(train_data_select['timestamp'], format='%Y-%m-%d'))
```

```python
#Dropping time stamp column
train_data_select.drop('timestamp', axis = 1, inplace=True)
```

```python
#build year
train_data_select = train_data_select.assign(BD_Year = (train_data_select['build_year'].astype(int)))

#modify BD_Year to only take values with 4 digit years
train_data_select = train_data_select.assign(BD_Year_2 = np.where(train_data_select['BD_Year'].astype(str).str.len() == 4,train_data_select['BD_Year'],0  ) )
```

```python
#Replicating the above code using dictionary - 
#Only finding unique values where unique values are less or dtype is objects
columndict = dict(train_data_select.dtypes)
for k, v in columndict.items():
    if train_data_select[k].dtype == 'object' and train_data_select[k].nunique() < 5  :
        print 'Unique values for', k ,': \n', train_data_select[k].value_counts(dropna = False)  
```

```python
#Check for missing values - NaN or NA
def num_missing(x):
    return sum(x.isnull())
def num_missing_perc(x):
    return (sum(x.isnull())/len(x)) * 100

#Applying per column:
print "Missing values per column:"
print train_data_select.apply(num_missing, axis=0)
print "Percentage of missing per column:"
print train_data_select.apply(num_missing_perc, axis=0)
print train_data_select.dtypes
```

```python
#Do manually
train_data_select['floor'] = train_data_select['floor'].fillna(0)
```

```python
#Same thing as above but automated
missingval_col = dict(train_data_select.apply(num_missing, axis=0))

for k, v in missingval_col.items():
    if train_data_select[k].dtype in ['float64','int64']:
        train_data_select[k] = train_data_select[k].fillna(0)
```
